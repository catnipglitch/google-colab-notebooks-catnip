### Initialize SDK client

```python
from google import genai
from google.genai import types
from PIL import Image
from IPython.display import Image, Markdown



client = genai.Client(api_key=GOOGLE_API_KEY)
```

```python
MODEL_ID = "gemini-2.0-flash-preview-image-generation" # @param ["gemini-2.0-flash-preview-image-generation","gemini-2.0-flash-lite","gemini-2.0-flash","gemini-2.5-flash-preview-04-17","gemini-2.5-pro-exp-03-25"] {"allow-input":true, isTemplate: true}

img_text_prompt = "Show me how to bake a macaron with images." # @param {"type":"string"}

```

```python
response = client.models.generate_content(
   model=MODEL_ID,
   contents=(
       img_text_prompt
   ),
   config=types.GenerateContentConfig(
        response_modalities=["TEXT", "IMAGE"]
   ),
)


for part in response.candidates[0].content.parts:
  if part.text is not None:
    display(Markdown(part.text))
  elif part.inline_data is not None:
    mime = part.inline_data.mime_type
    print(mime)
    data = part.inline_data.data
    display(Image(data=data))
```
